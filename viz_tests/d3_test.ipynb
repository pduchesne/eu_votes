```python
from IPython.display import display, Javascript, HTML
import json
display(Javascript("require.undef('d3'); require.config({paths: {d3: 'https://d3js.org/d3.v5'}});"))

display(Javascript(filename="d3_utils.js"))

def scatter(data, field_x, field_y, field_class, label_fn):
    display(Javascript("""
        (function(element){
            require(['d3_utils'], function(d3_utils) {
                d3_utils.scatter(element.get(0), %s, '%s', '%s', '%s', %s);
            });
        })(element);
    """ % (json.dumps(data), field_x, field_y, field_class, label_fn)))
```

```python
import pandas   
data = pandas.DataFrame([{'a':1, 'b':2}, {'a':1, 'b':1}, {'a':0, 'b':1.5}])    

scatter(data.to_dict('records'), 'a', 'b')


```
